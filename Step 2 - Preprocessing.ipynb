```python
import pandas as pd
import spacy
import banglanltk as bn
import re
```

```python
# Load the data
df = pd.read_csv('data/Bengalihatespeech.csv')
df.head()
```

```python
df.shape
```

```python
def remove_emojis(text):
    # Define a regex pattern for emojis
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text).strip()

df['sentence'] = df['sentence'].apply(remove_emojis)
```

```python
# Function to remove < br /> tags
def remove_br_tag(text):
    br_pattern = re.compile(r'<\s*br\s*/?>,*,?')
    return br_pattern.sub('', text)

df['sentence'] = df['sentence'].apply(remove_br_tag)
```

```python
# Function to remove <a href="...">...</a> tags
def remove_a_tag(text):
    a_pattern = re.compile(r'<a\s*.*?\s*>.*?</a>')
    return a_pattern.sub('', text)

df['sentence'] = df['sentence'].apply(remove_a_tag)
```

```python
# Function to remove English Characters and numbers
def remove_eng_char(text):
    pattern = re.compile(r'[a-zA-Z0-9]')
    return pattern.sub('', text)

df['sentence'] = df['sentence'].apply(remove_eng_char)
```

```python
# Function to remove Bangla numbers
def remove_bangla_num(text):
    pattern = re.compile(r'[০-৯]')
    return pattern.sub('', text)

df['sentence'] = df['sentence'].apply(remove_bangla_num)
```

```python
# Function to remove English punctuation marks
def remove_eng_punc(text):
    pattern = re.compile(r'[,\/|।?]')
    return pattern.sub(' ', text)

df['sentence'] = df['sentence'].apply(remove_eng_punc)
```

```python
nlp = spacy.blank("bn")
```

```python
def preprocess(text):
    # remove stop words 
    doc = nlp(text)
    
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.text)
    
    return " ".join(filtered_tokens) 

df['sentence'] = df['sentence'].apply(preprocess) 
```

```python
import BnLemma as lm
```

```python
bl = lm.Lemmatizer()

for i, sen in enumerate(df['sentence']):
    df.loc[i, 'sentence'] = bl.lemma(sen)
```

## Removing sentences with size less than 4 and greater than 200

```python
df = df[(df['sentence'].str.len()<200)&(df['sentence'].str.len()>4)]
```

```python
df.shape
```

```python
# Function to remove extra space
def remove_space(text):
    pattern = re.compile(r'\s+')
    return pattern.sub(' ', text)

df['sentence'] = df['sentence'].apply(remove_space)
```
