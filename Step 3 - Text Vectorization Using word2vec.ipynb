```python
import pandas as pd
import numpy as np
import multiprocessing
from gensim.models import Word2Vec
```

```python
df =pd.read_csv('data/preprocessed_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
def text_to_word_list(text):
    text = text.split()
    return text
```

```python
df['token'] = df['sentence'].apply(lambda x: text_to_word_list(str(x)))
word2vecinput = [row for row in df.token]
```

```python
df.head()
```

```python
model = Word2Vec(word2vecinput, vector_size=300, window=20, min_count=5, sg=0, negative=3, workers=multiprocessing.cpu_count()-1)
```

```python
print(model.wv.index_to_key)
```

```python
print(len(model.wv.index_to_key))
```

```python
model.wv.vector_size
```

```python
def sent_vec(sent):
    vector_size = model.wv.vector_size
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    wv = model.wv
    ctr = 1
    for w in sent:
        if w in wv:
            ctr += 1
            wv_res += wv[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
df['vec'] = df['token'].apply(sent_vec)
```
